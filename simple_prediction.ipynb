```
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

```
diameters = [[7], [10], [15], [30], [45]] # cm
prices = [[8],[11], [16], [38.5], [52]] # R$
```

```
plt.figure()
plt.xlabel('diameter (cm)')
plt.ylabel('price (R$)')
plt.title('diameter x price')
plt.plot(diameters, prices, 'k.')
plt.axis([0, 60, 0, 60])
plt.grid(True)
plt.show()
```

```
from sklearn.linear_model import LinearRegression
```

```
X = np.array(diameters).reshape((-1, 1)) # cm
Y = np.array(prices).reshape((-1, 1)) # R$
```

```
model = LinearRegression()
type(model)
```

```
model.fit(X, Y.reshape((-1, 1)))
```

```
# y = a + bx
prediction = model.predict([[0], [20]])
print('A 20 cm diameter pizza costs: R$ %.2f ' % prediction[1])
```

```
print('coefficient: \n', model.coef_)
print('mean square error: %.2f' % np.mean((model.predict(X) - Y) ** 2))
print('variation score: %.2f' % model.score(X, Y))
```

```
plt.scatter(X, Y, color = 'black')
plt.plot(X, model.predict(X), color  = 'blue', linewidth = 3)
plt.xlabel('diameters')
plt.ylabel('prices')
plt.xticks(())
plt.yticks(())

plt.show()
```
